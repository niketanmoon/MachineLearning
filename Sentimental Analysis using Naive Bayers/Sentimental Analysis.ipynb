```python
train_dataset["Sentiment_2"] = [1 if i == "pos" else 0 for i in train_dataset.label]
```

```python
def get_sentiment(label):
    if label == "pos":
        return 1
    else:
        return 0
```

```python
train_dataset["Sentiment_3"] = train_dataset.label.apply(get_sentiment)
```
